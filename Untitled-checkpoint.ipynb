```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('foot_line_shoes.csv')
```

```python
print(data.head())
```

```python
features = data[['listing_price', 'sale_price', 'discount', 'rating', 'reviews']]
```

```python
# Standardize the data
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)
```

```python
# Use the Elbow Method to find the optimal number of clusters
sse = []
k_range = range(1, 11)
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(scaled_features)
    sse.append(kmeans.inertia_)

# Plot the results of the Elbow Method
plt.figure(figsize=(10, 6))
plt.plot(k_range, sse, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Sum of Squared Errors (SSE)')
plt.title('Elbow Method for Optimal Number of Clusters')
plt.show()
```

```python
# Choose the optimal number of clusters (e.g., 3)
optimal_k = 3
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
kmeans.fit(scaled_features)
```

```python
data['cluster'] = kmeans.labels_
```

```python
# Melihat distribusi foot shoes dalam setiap cluster
cluster_distribution = data['cluster'].value_counts()
print(cluster_distribution)
```
